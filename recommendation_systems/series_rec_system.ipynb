```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Datasets/TMDB_tv_dataset_v3.csv',nrows=10000,usecols=["id","name","original_language","tagline","genres","overview","created_by"])
df.isna().sum()
```

```python
df.head()
```

```python
df["genres"].fillna(" ",inplace=True)
df["overview"].fillna(" ",inplace=True)
df["tagline"].fillna(" ",inplace=True)
```

```python
df.isna().sum()
```

```python
df["genres"] = df["genres"].apply(lambda x:str(x).replace(","," "))
df["created_by"] = df["created_by"].apply(lambda x:str(x).replace(" ",""))
```

```python
df.head()
```

```python
df["tags"] =  df["original_language"] + " " + df["tagline"] + " " + df["genres"] + " " + df["overview"] + " " +df["created_by"]
```

```python
df.head()
```

```python
df.drop(["original_language","tagline","genres","overview","created_by"],axis=1,inplace=True)
df.head()
```

```python
from nltk import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
df["tags"] = df["tags"].apply(lambda x:x.lower())
df["tags"] = df["tags"].apply(stem)
```

```python
df["tags"][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words="english")
vectors = cv.fit_transform(df["tags"]).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
del vectors # Free up memory
```

```python
similarity
```

```python
def recommend(movie_id):
  idx = df[df["id"]==movie_id].index[0]
  sim_scores = similarity[idx]
  movie_ids = df["id"].to_list()
  print(movie_ids)
  pair = list(zip(movie_ids,sim_scores))
  pair = sorted(pair,key=lambda x:x[1],reverse=True)[1:6]
  pair = [i[0] for i in pair]
  return pair

for i in recommend(66732):
  print(df[df["id"]==i])

# recommend(1402)
```
